## Investigating products

---

The `products` and `categories` tables contain information to answer this question. We will use a combination of SQL queries and Pandas to answer the following questions:

1. What categories of products is the company selling?
- How many products per category does the catalog contain?
- Let's focus only on products that have not been discontinued => how many products per category?
- What are the most expensive 5 products (not discontinued)?
- How many units of each of these 5 products are there in stock?
- Construct a bar chart of the data with pandas.

### 1. What categories of products is the company selling?

```python
query = 'SELECT DISTINCT "CategoryName" FROM categories'

pd.read_sql(query, con=engine)
```

### 2 How many products per category does the catalog contain?


```python
# We can "rename" out tables to not have to call them each time

query = """
SELECT c."CategoryName", COUNT(p."ProductID") AS "ProdsPerCat"
FROM categories c
INNER JOIN products p
ON p."CategoryID" = c."CategoryID"
GROUP BY c."CategoryName"
"""

pd.read_sql(query, con=engine)
```

### 3 How many _not discontinued_ products are there per category?

```python
# If we not include the COUNT for ProductID, we'll get the following error message:
# column "p.ProductID" must appear in the GROUP BY clause or be used in an aggregate function

query = """
SELECT c."CategoryName", COUNT(p."ProductID") AS "ProdsPerCat"
FROM categories c
INNER JOIN products p
ON p."CategoryID" = c."CategoryID"
WHERE p."Discontinued"=0
GROUP BY c."CategoryName"
"""

pd.read_sql(query, con=engine)
```

### 4. What are the top five most expensive products (not discontinued)?

```python
query = """
SELECT p."ProductName", p."UnitPrice"
FROM products p
WHERE p."Discontinued"=0
ORDER BY p."UnitPrice" DESC
LIMIT 5
"""

pd.read_sql(query, con=engine)
```

### 5 How many units of each of these 5 products are there in stock?

```python
# We can use UnitPrice column as a filter without selecting it

query = """
SELECT p."ProductName", p."UnitsInStock"
FROM products p
WHERE p."Discontinued"=0
ORDER BY p."UnitPrice" DESC
LIMIT 5
"""

data = pd.read_sql(query, con=engine)
data
```

### 6 Use pandas to make a useful bar chart of the product data.

```python
# A:
data.plot('ProductName', 'UnitsInStock', kind='bar', legend=False)
plt.ylabel('UnitsInStock');
```

## Investigating orders

---

Now that we have a better understanding of products, let's start digging into orders.

- How many orders are there in total?
- How many orders per year?
- How many orders per quarter?
- Which country is receiving the most orders?
- Which country is receiving the least?
- What's the average shipping time (ShippedDate - OrderDate)?
- What customer is submitting the highest number of orders?
- What customer is generating the highest revenue (need to pd.merge with order_details)?

### 1 How many orders in total?

```python
query = """
SELECT * 
FROM orders
"""

pd.read_sql(query, con=engine).tail()
```

```python
query = """
SELECT COUNT(*) AS "Num_Orders"
FROM orders o
"""

pd.read_sql(query, con=engine)
```

### 2 How many orders per year?  

We'll use the SQL [`Extract`](https://www.tutorialspoint.com/sql/sql-date-functions.htm#function_extract) function here.

```python
query = """
SELECT *
FROM orders o
LIMIT 2
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT CAST(EXTRACT(YEAR FROM o."OrderDate") AS INT) AS "Year", COUNT(*) AS "Num_Orders"
FROM orders o
GROUP BY "Year"
ORDER BY "Year"
"""

pd.read_sql(query, con=engine)
```

### 3 How many orders per quarter?

```python
query = """
SELECT CAST(EXTRACT(YEAR FROM o."OrderDate") AS INT) AS "Year", 
CAST(EXTRACT(QUARTER FROM o."OrderDate") AS INT) AS "Quarter", 
COUNT(*) AS "Num_Orders"
FROM orders o
GROUP BY "Year", "Quarter"
ORDER BY "Year", "Quarter"
"""

orders_per_quarter = pd.read_sql(query, con=engine)
orders_per_quarter
```

### 4 Which country is receiving the most orders?

```python
query = """
SELECT o."ShipCountry",
COUNT(*) AS "Num_Orders"
FROM orders o
GROUP BY o."ShipCountry"
ORDER BY "Num_Orders" DESC
LIMIT 1
"""

pd.read_sql(query, con=engine)
```

### 5 Which country is receiving the least?

```python
query = """
SELECT o."ShipCountry",
COUNT(*) AS "Num_Orders"
FROM orders o
GROUP BY o."ShipCountry"
ORDER BY "Num_Orders" ASC
LIMIT 1
"""

pd.read_sql(query, con=engine)
```

### 6 What's the average shipping time (ShippedDate - OrderDate)?

```python
query = """
SELECT AVG(o."ShippedDate"-o."OrderDate") AS "Avg_Shipping_Time"
FROM orders o
"""

pd.read_sql(query, con=engine)
```

### 7 What customer is submitting the highest number of orders?

```python
query = """
SELECT *
FROM orders o
LIMIT 1
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT o."CustomerID", COUNT(*) AS "Num_Orders"
FROM orders o
GROUP BY o."CustomerID"
ORDER BY "Num_Orders" DESC
LIMIT 1
"""

pd.read_sql(query, con=engine)
```

### 8 What customer is generating the highest revenue?

```python
query = """
SELECT o."CustomerID", SUM(od."UnitPrice"*od."Quantity"-od."Discount") AS "Revenue"
FROM orders o
INNER JOIN order_details od
ON od."OrderID"=o."OrderID"
GROUP BY o."CustomerID"
ORDER BY "Revenue" DESC
LIMIT 1
"""

pd.read_sql(query, con=engine)
```
